```python
from tensorflow.keras.layers import Dense;
from tensorflow.keras.optimizers import Adam, SGD
from dataset import Dataset;
from hyperparameter import HyperParameter;
from hyperparameter_search import HyperParameterSearch;
    
dataset_vector_25 = Dataset("../preprocessed/row_size25_vector_size2000.csv")
dataset_vector_100 = Dataset("../preprocessed/row_size100_vector_size2000.csv")
```

```python

parameters = [
     HyperParameter(
         dataset_vector_100,
         lambda: (
             [
                Dense(256,input_dim=dataset_vector_100.getInDimension(), activation="sigmoid"),
                Dense(128, activation="sigmoid"),
                Dense(64, activation="sigmoid"),
                Dense(32, activation="sigmoid"),
                Dense(dataset_vector_100.getOutDimension(),activation="sigmoid")

             ]
         ),
         '[Dense(256,input_dim=dataset_vector_100.getInDimension(), activation="sigmoid"),Dense(128, activation="sigmoid"), Dense(64, activation="sigmoid"), Dense(32, activation="sigmoid"), Dense(dataset_vector_100.getOutDimension(),activation="sigmoid")]',
         loss="mse", 
         optimizer="sgd",
         epochs=30,
         batch_size=12,
     ),
    HyperParameter(
         dataset_vector_25,
         lambda: (
             [
                Dense(256,input_dim=dataset_vector_25.getInDimension(), activation="sigmoid"),
                Dense(128, activation="sigmoid"),
                Dense(64, activation="sigmoid"),
                Dense(32, activation="sigmoid"),
                Dense(dataset_vector_25.getOutDimension(),activation="sigmoid")

             ]
         ),
         '[Dense(256,input_dim=dataset_vector_25.getInDimension(), activation="sigmoid"),Dense(128, activation="sigmoid"), Dense(64, activation="sigmoid"), Dense(32, activation="sigmoid"), Dense(dataset_vector_25.getOutDimension(),activation="sigmoid")]',
         loss="mse", 
         optimizer="sgd",
         epochs=30,
         batch_size=12,
     ),
]
```

```python
str(parameters[0])
```

```python
search = HyperParameterSearch();
search.run(parameters);
```
