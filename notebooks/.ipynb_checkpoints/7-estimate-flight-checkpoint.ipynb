## Configure dataset of a single file in same format

```python
import pandas as pd
import os

files = os.listdir("../data/")
FILE_PATHS = "../data/";

def get_dataframe(file_name):
    return pd.read_csv(FILE_PATHS + file_name)

df = get_dataframe(files[0])
df.head()
```

```python
import random

def read_random_file(files, columns):
    file = files[random.randint(0, len(files) - 1)]
    if (file == ".DS_Store"):
        file = files[random.randint(0, len(files) - 1)]
    df = get_dataframe(file)
    try:
        assert(str(df.columns) == str(columns))
    except AssertionError as error:
        print(file + " failed");
    return df;

read_random_file(files, df.columns).head()
```

```python
def get_index_rows(df, start_index, amount):
    if (df.shape[0] - amount <= 1):
        return pd.DataFrame()
    return df.iloc[(start_index):(amount + (start_index + 1)), :]
```

```python
get_index_rows(df, 0, 25)
```

```python
def convert_to_table_row(df):
    try:
        row = [
            df["timestamp"].iloc[-1] - df["timestamp"].iloc[0], # time_change
            df["linear_position_x"].iloc[-1] - df["linear_position_x"].iloc[0], # x_change
            df["linear_position_y"].iloc[-1] - df["linear_position_y"].iloc[0], # y_change
            df["linear_position_z"].iloc[-1] - df["linear_position_z"].iloc[0], # z_change
            df["angular_position_phi"].iloc[-1] - df["angular_position_phi"].iloc[0], # phi_change
            df["angular_position_theta"].iloc[-1] - df["angular_position_theta"].iloc[0], # theta_change
            df["angular_position_psi"].iloc[-1] - df["angular_position_psi"].iloc[0] # psi_change
        ];

        row += df["accelerometer_reading_x"].iloc[:-1].tolist()
        row += df["accelerometer_reading_y"].iloc[:-1].tolist()
        row += df["accelerometer_reading_z"].iloc[:-1].tolist()
        row += df["gyroscope_reading_phi"].iloc[:-1].tolist()
        row += df["gyroscope_reading_theta"].iloc[:-1].tolist()
        row += df["gyroscope_reading_psi"].iloc[:-1].tolist()
        return row
    except IndexError as error:
        print(df + " failed");
        return False

convert_to_table_row(get_index_rows(df, 1, 1))
```

```python
def generate_string_names(name, amount):
    names = []
    for i in range(amount):
        names.append(name + "_" + str(i));
        
    return names;

def generate_columns(amount):
    columns = [
        "time_change",
        "x_change",
        "y_change",
        "z_change",
        "phi_change",
        "theta_change",
        "psi_change",
    ];
    columns += generate_string_names("accelerometer_reading_x", amount);
    columns += generate_string_names("accelerometer_reading_y", amount);
    columns += generate_string_names("accelerometer_reading_z", amount);
    columns += generate_string_names("gyroscope_reading_phi", amount);
    columns += generate_string_names("gyroscope_reading_theta", amount);
    columns += generate_string_names("gyroscope_reading_psi", amount);
    return columns;

generate_columns(1)
```

```python
def generate_table(df, vector_size):
    table = []
    
    for i in range(df.shape[0] - vector_size):
        rows = get_index_rows(df, i, vector_size)
        if (rows.empty):
            continue
            
        row = convert_to_table_row(rows)
        if (row == False):
            continue
        table.append(row)
        
    return pd.DataFrame(table, columns=generate_columns(vector_size))

df_flight = generate_table(df, 25)
df_flight
```

```python
from dataset import Dataset;
from plotter import Plotter, PlotData;
import numpy as np

def getFlightPrediction(df, position_changes):
    time_prediction = []
    index = 0;
    for change in position_changes:
        if (index == 0):
            time_prediction.append(change[0])
        else:
            time_prediction.append(
                time_prediction[index - 1] + change[0])
        index = index + 1
    return time_prediction

def getPlotData(df, target_name, prediction, title):
    time_target = []

    index = 0
    for change in df[target_name].values:
        if (index == 0):
            time_target.append(change)
        else:
            time_target.append(time_target[index - 1] + change)

        index = index + 1
    
    return PlotData(df, pd.DataFrame(time_target), pd.DataFrame(getFlightPrediction(df, prediction)), title)

plots = [
    getPlotData(df_flight, 'x_change', model.prediction_x, "x Flight Results"),
    getPlotData(df_flight, 'y_change', model.prediction_y, "y Flight Results"),
    getPlotData(df_flight, 'z_change', model.prediction_z, "z Flight Results"),
    getPlotData(df_flight, 'phi_change', model.prediction_phi, "phi Flight Results"),
    getPlotData(df_flight, 'theta_change', model.prediction_theta, "theta Flight Results"),
    getPlotData(df_flight, 'psi_change', model.prediction_psi, "psi Flight Results"),
]
```

```python
Plotter().plot_predictions(plots, False);
```

```python
plots = [
    getPlotData(df_flight, 'x_change', model.prediction_x, "x Flight Results, c=3.0"),
    getPlotData(df_flight, 'y_change', model.prediction_y, "y Flight Results, c=2.55"),
    getPlotData(df_flight, 'z_change', model.prediction_z, "z Flight Results, c=1.75"),
    getPlotData(df_flight, 'phi_change', model.prediction_phi, "phi Flight Results"),
    getPlotData(df_flight, 'theta_change', model.prediction_theta, "theta Flight Results"),
    getPlotData(df_flight, 'psi_change', model.prediction_psi, "psi Flight Results"),
]

plots[0].prediction = plots[0].prediction * 3.0;
plots[1].prediction = plots[1].prediction * 2.55;
plots[2].prediction = plots[2].prediction * 1.75;
```

```python
Plotter().plot_predictions(plots, False);
```
